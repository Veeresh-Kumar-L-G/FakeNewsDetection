```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import re
import string
```

```python
data_fake = pd.read_csv('Fake.csv')
data_true = pd.read_csv('True.csv')
```

```python
data_fake['class'] = 0
data_true['class'] = 1
```

```python
data_fake_manual_test = data_fake.tail(10)
for i in range(23480, 23470, -1):
    data_fake.drop([i], axis = 0, inplace = True)


data_true_manual_test = data_true.tail(10)
for i in range(21416, 21406, -1):
    data_true.drop([i], axis = 0, inplace = True)
```

```python
data_fake_manual_test['class'] = 0
data_true_manual_test['class'] = 1
```

```python
data_true_manual_test.head(10)
```

```python
data_merge = pd.concat([data_fake, data_true], axis = 0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(['title', 'subject','date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace = True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)  # Remove text inside square brackets
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # Remove URLs
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'[{}]'.format(re.escape(string.punctuation)), '', text)  # Remove punctuation
    text = re.sub(r'\n', ' ', text)  # Replace newlines with spaces
    text = re.sub(r'\w*\d\w*', '', text)  # Remove words with digits
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xy_train = vectorization.fit_transform(x_train)
xy_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(xy_train, y_train)
```

```python
pred_lr = LR.predict(xy_test)
```

```python
LR.score(xy_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xy_train, y_train)
```

```python
pred_dt = DT.predict(xy_test)
```

```python
DT.score(xy_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier(random_state = 0 )
GB.fit(xy_train, y_train)
```

```python
pred_gb = GB.predict(xy_test)
```

```python
GB.score(xy_test, y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier()
RF = RandomForestClassifier(random_state = 0 )
RF.fit(xy_train, y_train)
```

```python
pred_rf = RF.predict(xy_test)
```

```python
RF.score(xy_test, y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
def evaluate_model(model, x_test, y_test):
    predictions = model.predict(x_test)
    print(f"{model.__class__.__name__} Accuracy: {model.score(x_test, y_test)}")
    print(classification_report(y_test, predictions))

evaluate_model(LR, xy_test, y_test)
evaluate_model(DT, xy_test, y_test)
evaluate_model(GB, xy_test, y_test)
evaluate_model(RF, xy_test, y_test)

# Define output label function
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"
    else:
        return "Unknown"

# Function for manual testing
def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    
    pred_lr = LR.predict(new_xv_test)
    pred_dt = DT.predict(new_xv_test)
    pred_gb = GB.predict(new_xv_test)
    pred_rf = RF.predict(new_xv_test)
    
    print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(
        output_label(pred_lr[0]),
        output_label(pred_dt[0]),
        output_label(pred_gb[0]),
        output_label(pred_rf[0])
    ))

# Take user input and test
if __name__ == "__main__":
    news_input = input("Enter the news text: ")
    manual_testing(news_input)
```
